```python
df = pd.read_csv('filebaru.csv',
                na_values = ['Budi','-', 'n.a']) # budi, - , n.a diganti NaN
 #kemudian na diganti dengan berikut.
df = df.fillna({
    'Nama': 'Anonim',
    'Usia': 24,
    'Gaji': 5000000
})
df

# df.loc[df.index.repeat(3)] 
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.fillna(method = 'ffill', axis='columns') #sama dengan # df = df.fillna(method = 'ffill', axis=1)
df
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.fillna(method = 'bfill', axis='rows') 
df
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.fillna(0)  ##harus di 0 kan dulu soalnya kalau tulisan atau NaN tidak bisa di ganti angka
df['Gaji'] = df ['Gaji'].astype('int64')  ##untuk jadi angka tipenya
# df.dtypes
df
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.interpolate()  ##untuk mengisi nilai tengah antara data sebelum dan sesudah

# df.dtypes
df
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.interpolate().fillna({'Nama': 'ANONIM'})  ##interpolate untuk mengisi nilai tengah antara data sebelum dan sesudah. bisa digabung fillna
df = df.dropna(thresh = 1) ##untuk drop yang ga ada isi, #threshold=1 artinya minimal 3 data harus ada baru bisa keluar
# df.dtypes
df
```

```python
df = pd.read_csv('filebaru.csv',
                na_values = ['-', 'n.a'])
# forward filling
df = df.interpolate().fillna({'Nama': 'ANONIM'})  ##interpolate untuk mengisi nilai tengah antara data sebelum dan sesudah. bisa digabung fillna
df = df.dropna(subset = ['Nama', 'Usia']) ##dropna = untuk drop yang ga ada isi,#subset=yang Nama dan Usia ada isinya maka dimunculkan
# df.dtypes
df
```
