```python
plt.style.use('seaborn-whitegrid')
x =np.arange(10)
y =np.arange(10)
plt.plot(x,y)
# plt.style.available  #cek ada style ap aja

df = pd.DataFrame({
    'x': np.arange(10),
    'y': np.arange(10),
    'z': ['Pria', 'Wanita'] * 5 
})
df
plt.plot(df.x,df.y)
```

```python
sb.get_dataset_names()
df = sb.load_dataset('flights')
# df.isnull().sum()  ## untuk kosongin tabel, tinggal diisi dgn yg mau. atau lihat ada xolumns apa saja
df = df.pivot('month', 'year', 'passengers') ## untuk menjadikan, month row, year sebagai column
df = df.sum()  ## menjumlahkan seluruh column
df = df.to_frame() ## untuk menjadikan tampilan sprt dataframe
df
# df[1949]
sb.barplot(data=df, x=df.index, y=0, orient='v')  ##y=0 karena x=df.index
# sb.barplot(data=df, y=df.index, x=0, orient='h') ##untuk membalik x jadi horizontal
plt.show()
```

```python
sb.get_dataset_names()
df = sb.load_dataset('flights')
df = df.pivot('month', 'year', 'passengers')
# df
sb.heatmap(df, annot=True, fmt='d', cmap='hot_r') ## fmt='d' agar muncul annot nya rapih
# 'hot_r' untuk ngebalik 'hot'
# plt.imshow(df)

df = df.corr()
df
```
