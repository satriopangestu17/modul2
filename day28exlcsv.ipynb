# Data Pegawai PT ABC

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data_ptabc.csv',
delimiter = ','
# , header =2
                 , skiprows = 2
                 , engine = 'python'   ##
                 , skipfooter = 2
)
df.set_index('No.')
```

```python
df['Nama']
```

```python
df['Gaji'].describe()
```

```python
df[['Nama','Usia']]
```

```python
df['Gaji'].describe().loc['max']
```

```python
df['Gaji'].max()
```

```python
df['Gaji'].min()
```

```python
df['Gaji'].mean()
```

```python
df['Gaji'].sum()
```

```python
df['Gaji'].mode()
```

```python
df['Gaji'].median()
```

```python
df[df['Gaji'] == df['Gaji'].min()]
```

```python
df[df['Gaji'] == df['Gaji'].min()][['Nama','Usia']]
```

```python
df[['Nama','Usia']][df['Gaji'] == df['Gaji'].min()]
```

```python
df[df['Usia'] == df['Usia'].max()]
```

```python
df[df['Usia'] != 35]
```

```python
df[df['Usia'] > 25][df['Kota'] == 'Jakarta']
```

```python
df[(df['Usia'] > 25) & (df['Usia'] < 30)]
```

```python
df[(df['Usia'].between(25, 30))]  ##kalau between 25 dan 30nya masuk, conditional berlaku untuk semua bukan hanya csv
```
