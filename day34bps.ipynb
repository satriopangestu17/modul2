```python
import pandas as pd
import numpy as np
##di save as sebagai xls. kalau mau diimport sebagai xlsx harus diopen dulu lewat number baru di export jadi xlsx
df = pd.read_excel('indo_12.xls', header=3, skipfooter=2,na_values = ['-', 'n.a']) ## kalau ada yang - dan n.a maka diganti dengan fillna(0)*jadi nol
df = df.rename({'Unnamed: 0': 'Provinsi'}, axis=1)
# df = df.fillna(method:'ffill', axis='index' ) ##ga mesti kalau ga mau dijadiin jadi 0
df = df.fillna(0)
df
df['Provinsi']
# len(df['Provinsi'])
# prov = []
# for i in range(0,len(df['Provinsi'])):
#     prov.append(df['Provinsi'][i])
# prov  ##ga bisa kalau pakai ini jadi harus dinamain satu2 ticksnya.


# df[1971]
# # len(df)
tahun = df.columns[1::].tolist()
tahun
stahun = pd.Series(tahun)
stahun
# interpolate(method='linear')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d



fig = plt.figure('3d', figsize=(16,8))
p = plt.subplot(111, projection='3d')

for i in range(len(tahun)):
    x = np.arange(len(df['Provinsi'])) #data x
    y = [i] * len(df['Provinsi']) #data y
    z = np.zeros(len(df['Provinsi']))
    dx = np.ones(len(df['Provinsi']))
    dy = np.ones(len(df['Provinsi']))
    dz = df[tahun[i]]
    p.bar3d(x, y, z, dx, dy, dz)
p.set_xlabel('sumbu X')
p.set_ylabel('sumbu Y')
p.set_xticks(np.arange(len(df['Provinsi'])))
p.set_xticklabels(df['Provinsi'], rotation='90')
p.set_yticks(np.arange(len(df['Provinsi'])))
p.set_yticklabels(tahun, rotation='45')
p.set_zlabel('Populasi')
plt.tight_layout()
plt.show()
```
