```python
import pandas as pd
import numpy as np

dftlkm = pd.read_csv('TLKM.JK.csv', parse_dates = ['Date'],   ##untuk menjadikan tipe data dates
                    index_col = ['Date'])
dftlkm = dftlkm.sort_index()
dftlkm = dftlkm.resample('D').sum()  ##hari('D') libur jadi muncul
dftlkm.iloc[:, :5] = dftlkm.iloc[:, :5].replace(0, np.NaN)
dftlkm = dftlkm.fillna(method='ffill')
dftlkm
```
