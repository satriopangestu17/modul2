```python
import pandas as pd
import numpy as np
##merge = join
df1 = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Budi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Caca', 'fisika': 92, 'biologi': 82},])
df2 = pd.DataFrame([
    {'nama': 'Andi', 'math': 90, 'kimia': 95},
    {'nama': 'Budi', 'math': 80, 'kimia': 98},
    {'nama': 'Caca', 'math': 88, 'kimia': 90},
    {'nama': 'Deni', 'math': 88, 'kimia': 90},
])
## outer join agar semua keluar termasuk Deni
dfx = pd.merge(df1, df2, on='nama', how='outer')
dfx
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'mapel': ['biologi', 'biologi', 'biologi', 'fisika', 'fisika', 'fisika'],
    'test': [1,2,3,1,2,3],
    'Andi': [78,80,82,90,95,92],
    'Budi': [88,89,72,80,85,72]
})
df = df.set_index(['test', 'mapel']).unstack() 
# df = df.set_index(['mapel', 'test']).unstack()##unstack bikin jadi single index. multi indexnya jadi columns karena yang di 0 nya test
df = df.swaplevel(0,1,1).sort_index(1)   ##swaplevel agar mapel bilogi fisika diatas nama andi dan budi, 0 dan 1 untuk yg mana yg mau d swap dan 1 yang diakhir untuk axis. #sort agar tersort tidak keluar banyak
df
```
