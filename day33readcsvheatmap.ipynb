```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im

df = pd.read_csv('atlit.csv')
dfCorr = df.corr()
##jadi kurangi makan
plt.imshow(dfCorr, cmap='hot_r')

plt.colorbar()
col = list(map(lambda x: x, dfCorr.columns.tolist()))
plt.xticks(np.arange(len(dfCorr.index)),dfCorr.index.tolist(), rotation=45)
plt.yticks(np.arange(len(dfCorr.index)),dfCorr.index.tolist())
for x in range(len(col)):
    for y in range(len(col)):
        plt.text(y, x, round(dfCorr.iloc[x, y],3), color='blue',fontsize=10, ha='center', va='center')
# plt.xlim(-.5,4.5)
plt.xlim(-.5,4.5)
plt.ylim(-.5,4.5)
plt.show()
# col
```
