```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.bar(df.x, df.y, color=['r','g','b','k','yellow'], edgecolor='black', 
        linewidth=5)
plt.barh(df.x, df.y, color=['r','g','b','k','yellow'], edgecolor='black', 
        linewidth=5)
plt.yticks(np.arange(0,12,1))
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size':30, 'color': 'darkgreen'
})
# for i in df['y']:
#     plt.text(df[df['y'] == i]['x'], i, i, fontdict={'size': 20})  #plt.text(x,y,value)
for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.1, i+1, i, fontdict={'size': 20}) 
plt.xlabel('ini x', fontsize=20)
plt.ylabel('ini y', fontsize=20)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.bar(df.x, df.y, color=['r','g','b','k','lightgreen'], edgecolor='black', 
        linewidth=5)
plt.bar(df.x, df.z, color=['yellow'], bottom= df.y, yerr = df.describe()['y'].loc['std'], edgecolor='black', 
        linewidth=5)  #bottom jadi diatas nilai tersebut gambar grafik. #yerr untuk perkiraan eror
plt.yticks(np.arange(0,12,1))
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size':30, 'color': 'darkgreen'
})
# for i in df['y']:
#     plt.text(df[df['y'] == i]['x'], i, i, fontdict={'size': 20})  #plt.text(x,y,value)
for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.1, i+1, i, fontdict={'size': 20}) 
plt.xlabel('ini x', fontsize=20)
plt.ylabel('ini y', fontsize=20)
plt.show()
df.describe()['y'].loc['std'] ##std untuk 'y'
```

```python
#selain bar dan plot ada
#SCATTER
##pakai zorder untuk nentuin order gambar atas bawah. yang lebih besar yang di atas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.plot([1,2,3], [4,5,6], 'r', markersize = 20, zorder=0)
plt.bar([1,2,3], [4,5,6], color='lightgreen', zorder=-1)
plt.scatter(
[1,2,3], [4,5,6], marker='X', s=300, color='yellow', edgecolor='k', zorder=1)
plt.show
```

```python
##HISTOGRAM
usia = [22,21,31,42,32,45,65,55,43,22,21,34,43,44]
usia2 = [33,31,41,21,22,25,25,25,43,22,21,34,43,44]
par = [15,20,25,30,35,40,45,50,55,60,65]
plt.hist([usia, usia2], par, histtype='step', rwidth=10)
plt.xticks(np.arange(15,66,5))
plt.show()
```
